```python
# ------------ FUNCTIONS ---------------

def merge(array, left, middle, right, param):
    n1 = middle - left + 1
    n2 = right - middle
    
    array1 = []
    array2 = []

    for i in range(n1): 
        array1.append(array[left + i])
        
    for j in range(n2):
        array2.append(array[middle + 1 + j])
  
    i = 0
    j = 0
    k = left
    
    while i < n1 and j < n2:
        if array1[i][param] <= array2[j][param]:
            array[k] = array1[i]
            i = i + 1
        else: 
            array[k] = array2[j]
            j = j + 1
        k = k + 1
        
        
    while i < n1:
        array[k] = array1[i] 
        i = i + 1
        k = k + 1
  
    while j < n2:
        array[k] = array2[j] 
        j = j + 1
        k = k + 1
        
    return array

def mergeSortAction(arr, left, right, param):
    if left < right: 
        middle = (left+(right-1))//2

        mergeSortAction(arr, left, middle, param) 
        mergeSortAction(arr, middle+1, right, param) 
        merge(arr, left, middle, right, param) 
    return arr

def mergeSort(arr, param):
    return mergeSortAction(arr, 0, len(arr)-1, param)
```
